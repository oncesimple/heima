# 微博热搜评论
热搜url = https://s.weibo.com/top/summary?Refer=top_hot&topnav=1&wvr=6

```python
import requests
import pandas as pd
from lxml import etree
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/79.0.3945.130 Safari/537.36 '
}
def reque(url):
    return requests.get(url,headers=headers).text

url = 'https://s.weibo.com/top/summary?Refer=top_hot&topnav=1&wvr=6'
data = reque(url)
```

## 把HTML读成pandas文件

```python
fdf = pd.read_html(data)
fdf = fdf[0]
fdf['序号'][0] = '0.0'
fdf = fdf.set_index('序号')
fdf.loc[:, "关键词"] = fdf["关键词"].str.replace("\s\d*", "")
fdf
```

## 添加一行热评链接

```python
# hh = fdf['关键字'].to_list()
fdf.loc[:, "链接"] = 'https://s.weibo.com/weibo?q=' + fdf['关键词']
fdf
```

### 搜索页进去后进子页
xpath = '//*[@id="pl_feedlist_index"]//div[1]/div[2]/p[2]/a[1]/@href'  # 主页里面子页的url

```python
def html_xpath(html,xps):
    html = etree.HTML(html)
    return html.xpath(xps)
```

```python
url = 'https://s.weibo.com/weibo?q=美媒称约150名沙特王室成员感染新冠'
data1 = reque(url)
```
